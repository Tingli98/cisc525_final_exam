## 3. How many aircraft types (use `aircraft_type`) are there?

````python
# Count the type of aircraft with the distinct function
aircraft_type = airline.planes.distinct('aircraft_type')
print(len(aircraft_type), 'types of aircraft')
aircraft_type
````

## 4. How many types of planes `CESSNA` (use `manufacturer` and `aircaft_type`, made over the years?

````python
# Calculate the number of the aircaft type for which the manufacture is CESSNA
CESSNA_plane_type = airline.planes.distinct("aircraft_type",{'manufacturer':'CESSNA'})
print(len(CESSNA_plane_type), 'types of aircraft CESSNA made')
CESSNA_plane_type
````

## 5. List aircraft types made by each manufacturer

````python
# 1. Find all of the manufacture
# 2. Iterate the database, find the aircraft types for every manufature
manufaturers = airline.planes.distinct('manufacturer')
for manufaturer in manufaturers:
    aircraft_type = airline.planes.distinct("aircraft_type",{'manufacturer':manufaturer})
    print('**', manufaturer, ': ', aircraft_type, '\n')
````

## 6. List number of planes make by each manufacturer

````python
# 1. Get the manufaturers
# 2. Iterate the database and count the plane numbers for every manufacturer
manufacturer = airline.planes.distinct('manufacturer')
manufacture_plane_num = []
for manufaturer in manufaturers:
    count = len(list(airline.planes.find({'manufacturer':manufaturer})))
    print('**', manufaturer, ': ', count, '\n')
    manufacture_plane_num.append([manufaturer, count])
````

````python
# Sort the (manufatucre, number) dataset for finding the most and the least number of planes made by the manufacturer
manufacture_plane_num.sort(key = lambda x: x[1])
````

## 7. Which manufacturer makes the `least` planes?

````python
# 1. The smallest number is 1, so print the manufacture when the plane number is 1
for num in manufacture_plane_num:
    if num[1] == 1:
        print(num, '\n')
````

## 8. Which manufacturer makes the `most` planes?

````python
# The list is sorted in ascending order, so the last element will be the manufacture which makes the most planes
manufacture_plane_num[-1]
````

## 9. Which state has the `most` airports?

````python
# 1. Get the states
# 2. Iterate the database and count the unique airports number for every state
states = airline.airports.distinct('state')
state_airports_num = []
for state in states:
    count = len(airline.airports.distinct('airport',{'state':state}))
    print('**', state, ': ', count, '\n')
    state_airports_num.append([state, count])
````

## 10. Which state has the `least` airports?

````python
# Sort the state_airports_num dataset for finding the most and the least number of airports in state
state_airports_num.sort(key = lambda x: x[1])
state_airports_num
````

````python
# The least number of airport is 1, so print out the airport which the number is one
for state in state_airports_num:
    if state[1] == 1:
        print(state, '\n')
````

## 11. Which state has the `second` largest number of airports?

````python
print('The state has the second larget number of aiports is : ', state_airports_num[-2])
````

## 12. List all the years in the `planes` collection. Exclude (0 and 'None')

````python
years = airline.planes.distinct('year')
filtered_years = list(filter(lambda num: ((num != 0) & (num != 'None')) , years))
````

## 13. Which year produced the `most` planes?

````python
# Count the plane number for each year, and save the year, count data to the sorted_data list
sorted_year = []
for year in filtered_years:
    plane_num = len(list(airline.planes.find({'year':year})))
    sorted_year.append([year,plane_num])
sorted_year.sort(key = lambda x: x[1])
sorted_year
````

````python
# year with most planes
print('year produce the most planes: ',sorted_year[-1])
````

## 14. Which year produced the `least` planes?

````python
# Print out the years when the plane number equal to 1
for year in sorted_year:
    if year[1] == 1:
        print(year, '\n')
````

## 16. List engine types made by each manufacturer

````python
# Get the unique manufacturer and get the unique engine type for every manufaturer with distinct function
manufaturers = airline.planes.distinct('manufacturer')
for manufaturer in manufaturers:
    engine_type = airline.planes.distinct("engine_type",{'manufacturer':manufaturer})
    print('**', manufaturer, ': ', engine_type, '\n')
````

## 17. List planes made by aircraft types

````python
# Get the unique type of aircaft
# Iterate the database with the aircaft type and list the plane info
aircraft_type = airline.planes.distinct('aircraft_type')
for aircraft_t in aircraft_type:
    planes = list(airline.planes.find({'aircraft_type':aircraft_t}))
    print('*******', aircraft_t, ': *******\n')
    for plane in planes:
        print('*',plane, '\n')
````

## 18. List planes made by engine type

````python
# Get the unique type of engine
# Iterate the database with the engine and list the plane info
engine_type = airline.planes.distinct('engine_type')
print(engine_type)
for engine in engine_type:
    planes = list(airline.planes.find({'engine_type':engine}))
    print('*******', engine, ': *******\n')
    for plane in planes:
        print('*',plane, '\n')
````
